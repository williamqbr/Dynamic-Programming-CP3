Projeto que modela redes de metrô como grafos ponderados e utiliza recursão com memoização para encontrar caminhos ótimos.

```python
import time
import tracemalloc
import functools

def fator_horario(hora):
    if 5 <= hora < 7:
        return 0.6
    elif 7 <= hora < 9:
        return 1.5
    elif 9 <= hora < 17:
        return 1.0
    elif 17 <= hora < 20:
        return 2.0
    else:
        return 1.2


def criar_menor_custo(grafo):
    @functools.lru_cache(maxsize=None)
    def menor_custo(origem, destino, hora, visitados=frozenset()):
        if origem == destino:
            return 0

        melhor = float('inf')

        for vizinho, peso in grafo.get(origem, []):
            if vizinho not in visitados:
                custo = fator_horario(hora) * peso + menor_custo(
                    vizinho,
                    destino,
                    hora,
                    visitados | {origem}
                )
                melhor = min(melhor, custo)

        return melhor

    return menor_custo


def criar_menor_custo_sem_cache(grafo):
    def menor_custo(origem, destino, hora, visitados=frozenset()):
        if origem == destino:
            return 0

        melhor = float('inf')

        for vizinho, peso in grafo.get(origem, []):
            if vizinho not in visitados:
                custo = fator_horario(hora) * peso + menor_custo(
                    vizinho,
                    destino,
                    hora,
                    visitados | {origem}
                )
                melhor = min(melhor, custo)

        return melhor

    return menor_custo


def caminho_mais_longo(grafo, origem, destino, visitados=None):
    if visitados is None:
        visitados = set()

    if origem == destino:
        return 0

    visitados.add(origem)
    pior = float('-inf')

    for vizinho, peso in grafo.get(origem, []):
        if vizinho not in visitados:
            custo = peso + caminho_mais_longo(
                grafo,
                vizinho,
                destino,
                visitados.copy()
            )
            pior = max(pior, custo)

    return pior

def medir(func, *args):
    tracemalloc.start()
    t0 = time.perf_counter()

    resultado = func(*args)

    t1 = time.perf_counter()
    mem = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    return resultado, t1 - t0, mem[1] / 1024


# GRAFOS
grafo_sp = {
    "Tucuruvi": [("Parada Inglesa", 3)],
    "Parada Inglesa": [("Tucuruvi", 3), ("Jardim São Paulo", 3)],
    "Jardim São Paulo": [("Parada Inglesa", 3), ("Santana", 4)],
    "Santana": [("Jardim São Paulo", 4), ("Carandiru", 3)],
    "Carandiru": [("Santana", 3), ("Portuguesa-Tietê", 2)],
    "Portuguesa-Tietê": [("Carandiru", 2), ("Armênia", 3)],
    "Armênia": [("Portuguesa-Tietê", 3), ("Tiradentes", 2)],
    "Tiradentes": [("Armênia", 2), ("Luz", 2)],

    "Luz": [("Tiradentes", 2), ("Sé", 4), ("Brás", 5)],

    "Sé": [("Luz", 4), ("São Bento", 2), ("Ana Rosa", 6)],
    "São Bento": [("Sé", 2), ("Anhangabaú", 2)],
    "Anhangabaú": [("São Bento", 2), ("República", 2)],
    "República": [("Anhangabaú", 2), ("Santa Cecília", 3)],
    "Santa Cecília": [("República", 3), ("Barra Funda", 5)],

    "Ana Rosa": [("Sé", 6), ("Paraíso", 2), ("Chácara Klabin", 5)],
    "Paraíso": [("Ana Rosa", 2), ("Vergueiro", 2)],
    "Vergueiro": [("Paraíso", 2), ("São Joaquim", 2)],
    "São Joaquim": [("Vergueiro", 2), ("Liberdade", 2)],
    "Liberdade": [("São Joaquim", 2), ("Sé", 3)],

    "Chácara Klabin": [("Ana Rosa", 5), ("Santos-Imigrantes", 3)],
    "Santos-Imigrantes": [("Chácara Klabin", 3), ("Alto do Ipiranga", 3)],
    "Alto do Ipiranga": [("Santos-Imigrantes", 3), ("Capão Redondo", 10)],
    "Capão Redondo": [("Alto do Ipiranga", 10)]
}


grafo_sf = {
    "Dublin/Pleasanton": [("West Dublin", 4), ("Castro Valley", 6)],
    "West Dublin": [("Dublin/Pleasanton", 4), ("Castro Valley", 5)],
    "Castro Valley": [("West Dublin", 5), ("Bay Fair", 4)],

    "Bay Fair": [("Castro Valley", 4), ("San Leandro", 3), ("Hayward", 6)],

    "San Leandro": [("Bay Fair", 3), ("Coliseum", 4)],
    "Hayward": [("Bay Fair", 6), ("South Hayward", 3)],

    "South Hayward": [("Hayward", 3), ("Union City", 4)],
    "Union City": [("South Hayward", 4), ("Fremont", 5)],
    "Fremont": [("Union City", 5), ("Coliseum", 10)],

    "Coliseum": [("San Leandro", 4), ("Fremont", 10), ("Fruitvale", 3)],
    "Fruitvale": [("Coliseum", 3), ("Lake Merritt", 2)],
    "Lake Merritt": [("Fruitvale", 2), ("West Oakland", 5)],

    "West Oakland": [("Lake Merritt", 5), ("Embarcadero", 6)],
    "Embarcadero": [("West Oakland", 6), ("Montgomery", 2)],
    "Montgomery": [("Embarcadero", 2), ("Powell", 2)],
    "Powell": [("Montgomery", 2), ("Civic Center", 2)],
    "Civic Center": [("Powell", 2), ("16th St Mission", 3)],
    "16th St Mission": [("Civic Center", 3), ("24th St Mission", 3)],
    "24th St Mission": [("16th St Mission", 3), ("Glen Park", 4)],
    "Glen Park": [("24th St Mission", 4), ("Balboa Park", 3)],
    "Balboa Park": [("Glen Park", 3), ("Daly City", 2)],

    "Daly City": [("Balboa Park", 2)]
}


grafo_bj = {
    "Sihui East": [("Sihui", 2)],
    "Sihui": [("Sihui East", 2), ("Guomao", 4)],
    "Guomao": [("Sihui", 4), ("Yonganli", 2), ("Jinsong", 5)],

    "Yonganli": [("Guomao", 2), ("Jianguomen", 3)],
    "Jianguomen": [("Yonganli", 3), ("Beijing Railway", 2)],

    "Beijing Railway": [("Jianguomen", 2), ("Chongwenmen", 2)],
    "Chongwenmen": [("Beijing Railway", 2), ("Qianmen", 3)],

    "Qianmen": [("Chongwenmen", 3), ("Hepingmen", 2)],
    "Hepingmen": [("Qianmen", 2), ("Xidan", 3)],

    "Xidan": [("Hepingmen", 3), ("Fuxingmen", 4), ("Xizhimen", 6)],

    "Fuxingmen": [("Xidan", 4), ("Chegongzhuang", 3)],
    "Chegongzhuang": [("Fuxingmen", 3), ("Xizhimen", 4)],

    "Xizhimen": [("Chegongzhuang", 4), ("Xidan", 6)],

    # loop alternativo
    "Jinsong": [("Guomao", 5), ("Shuangjing", 3)],
    "Shuangjing": [("Jinsong", 3), ("Guomao", 6)]
}

# EXECUÇÃO

def executar(nome, grafo, origem, destino, hora):
    print(f"\n===== {nome} =====")

    com_cache = criar_menor_custo(grafo)
    sem_cache = criar_menor_custo_sem_cache(grafo)

    r1, t1, m1 = medir(com_cache, origem, destino, hora)
    r2, t2, m2 = medir(sem_cache, origem, destino, hora)
    r3, t3, m3 = medir(caminho_mais_longo, grafo, origem, destino)

    print(f"Menor (memo): {r1:.2f} | {t1:.6f}s | {m1:.2f} KB")
    print(f"Menor (sem memo): {r2:.2f} | {t2:.6f}s | {m2:.2f} KB")
    print(f"Maior caminho: {r3:.2f} | {t3:.6f}s | {m3:.2f} KB")

hora = 18

executar("São Paulo", grafo_sp, "Tucuruvi", "Capão Redondo", hora)
executar("San Francisco", grafo_sf, "Dublin/Pleasanton", "Daly City", hora)
executar("Beijing", grafo_bj, "Sihui East", "Xizhimen", hora)
```
